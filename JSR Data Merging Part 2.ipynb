## Step 5. Merge orders_products_combined dataframe with cleaned products dataframe.

```python
# Drop unnamed columns in df_prods
df_prods = df_prods.drop(columns = ['Unnamed: 0'])
```

```python
# Merge df_orders_products_combined and df_prods
df_merged = df_prods.merge(df_orders_products_combined, on = 'product_id', indicator = True)
```
